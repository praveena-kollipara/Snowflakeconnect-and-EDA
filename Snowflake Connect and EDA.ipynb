```python
#Identifying Connecting flights

query = """
WITH UNIQUE_COMBI AS (
  SELECT DISTINCT FLIGHTNUMBER, STATION, DESTINATION
  FROM LOCAL_DATABASE.ORAAUE.UNT_BAGS
  GROUP BY FLIGHTNUMBER, STATION, DESTINATION
)
SELECT UC1.FLIGHTNUMBER, UC1.STATION AS STATION1, UC1.DESTINATION AS DESTINATION1,
       UC2.STATION AS STATION2, UC2.DESTINATION AS DESTINATION2
FROM UNIQUE_COMBI UC1
INNER JOIN UNIQUE_COMBI UC2 ON UC1.FLIGHTNUMBER = UC2.FLIGHTNUMBER
WHERE (UC1.STATION = UC2.DESTINATION)
ORDER BY UC1.FLIGHTNUMBER;
"""
cur.execute(query)
results = cur.fetchall()
```

```python
column_names = [desc[0] for desc in cur.description]
```

```python
df_connectingflights=pd.DataFrame(results, columns=column_names)
```
